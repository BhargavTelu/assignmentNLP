```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from textblob import TextBlob
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("Womens Clothing E-Commerce Reviews.csv")
```

```python
data_women=data_women.dropna()
```

```python
def preprocess(x):
    x=re.sub("[^a-zA-Z ]","",x)
    x=x.lower()
    x=x.split()
    x=[word for word in x if word not in set(stopwords.words("english"))]
    x=[PorterStemmer().stem(word) for word in x]
    x=" ".join(x)
    return x
```

```python
data_women["Review Text"]=data_women["Review Text"].apply(preprocess)
```

### polarity of text

```python
def getpolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
data_women["polarity"]=data_women["Review Text"].apply(getpolarity)
```

```python
def getanlaysis(score):
     if score < 0:
        return -1
     elif score == 0:
        return 0
     else:
        return 1
```

```python
data_women["sentiment"]=data_women["polarity"].apply(getanlaysis)
```

```python
data_women=data_women.drop(columns="polarity")
```

```python
data_women
```

### Count vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv=CountVectorizer()
X=cv.fit_transform(data_women["Review Text"].values).toarray()
X=pd.DataFrame(X,columns=cv.get_feature_names())
```

```python
X
```

# Decision Tree

```python
x_train,x_test,y_train,y_test=train_test_split(X,data_women.sentiment,test_size=0.3,random_state=10)
```

```python
model_decision=DecisionTreeClassifier(random_state=10)
model_decision.fit(x_train,y_train)
pred=model_decision.predict(x_test)
print(model_decision," Accuracy : ",metrics.accuracy_score(pred,y_test))
```

```python
model_Logistic=LogisticRegression(random_state=10)
model_Logistic.fit(x_train,y_train)
pred=model_Logistic.predict(x_test)
print(model_Logistic," Accuracy : ",metrics.accuracy_score(pred,y_test))
```
